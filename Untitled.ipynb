        <h1 class="c45 c25"><span class="c2">1. Introduction</span></h1><p class="c23"><span class="c2">Internet of Things (IoT), as a technological revolution (Madakam et al. 2015), is considered as the most hyped concept in the IT world (Kosmatos et al. 2011). By 2020, from the experts&#39; estimation, IoT will consist of about 30 billion objects (Tiwary et al. 2018). The advent of novel technologies and techniques in IoT is growing rapidly (Gim et al. 2016) so, getting the intellectual structure of the IoT field, identifying effective areas and relationships between these areas can be helpful for research centers, marketing, industries, and organizations. IoT documents have been studied from different aspects in academic fields (Yan et al. 2015). For example, Whitmore et al.(Whitmore et al. 2015) have classified 127 IoT papers based on the following major categories: technology, applications, challenges, business models, future directions, and overview/survey. Some of these top-level categories were further broken down into sub-categories (like technology has broken down to hardware, software, and architecture) and some of the sub-categories were broken into sub-sub-categories. Ray (Ray 2018) has mentioned application domains of IoT such as application development, device management, system management, heterogeneity management, data management, analytics, deployment, monitoring, visualization, and finally, research purposes. In the following, pointed out IoT domains(RFID, service-oriented architecture, wireless sensor network, supply chain management, industry, healthcare, smart city, logistics, connected living, big data, cloud computing, social computing, and security ) and surveyed IoT papers to understanding how IoT has been applied on sub-domains using particular architecture. Tiwary et al.(Tiwary et al. 2018) have studied the concept of IoT and architectures of that, and several of the advantages and disadvantages of the IoT are introduced. Zhong and Ge (Zhong and Ge 2018) have identified different models as well as their architectures according to the implementation approach, from published papers in the IoT domain between 2008 and 2016. RFID&#39;s documents have analyzed in terms of the number of documents per year, subject areas of the documents, and the number of documents in each country. They also examined the movement and growth of the IoT in three regions of Europe, North America and the Asia Pacific by providing benchmarks for each of the areas, and eventually sorted out countries based on papers that were published in Scopus.</span></p><p class="c23"><span class="c2">The mentioned papers have used traditional methods for identifying trends, that require expert&rsquo;s judgment and also can not specify effective areas and relationships between these areas. On the other hand, by growing tides of information generated in different sections, it seems that studying and following the trends of science with traditional methods is impossible. Therefore, the use of automatic methods and algorithms like co-word analysis in joint with machine learning algorithms to study the context of the generated data can be helpful. The co-word analysis is based on the diversity theory and can sufficiently cover the document&rsquo;s content. The method calculates the intensity of words communications using their co-occurrence in documents (Coulter et al. 1996). This approach reduces the concept dimension to graphs and discloses the governed relationships between the contents effectively (Coulter et al. 1998). Yan et al. (Yan et al. 2015) have analyzed the web of science(WOS) database papers from the period between 2004 and 2008 by co-word analysis method, and finally, they achieved seven clusters using the hierarchical clustering algorithm that represents the intellectual structure of IoT, including &lsquo;IoT and Security&rsquo;, &lsquo;Middleware&rsquo;, &lsquo;RFID&rsquo;, &lsquo;Internet&rsquo;, &lsquo;Cloud computing&rsquo;, &lsquo;Wireless sensor networks&rsquo; and &lsquo;6LoWPAN&rsquo;. </span></p><p class="c23"><span class="c2">However, despite the comprehensive studies on the intellectual structure of the IoT, the common point of these papers is that they have studied documents only from academic databases. On the other side, much information is available on the websites (Non-academic contents), and the volume of this information grows at a high rate every day. Also, due to the time-consuming procedure for the scientific papers getting accepted to publish on such databases, the contents available on blogs and websites would contain newer information. Therefore, investigating web page information is of great importance. As a result, in this research, the co-word analysis is applied on extracted documents from the Internet of Things blogs and websites, to approach this research gap. Another innovation in this research is that we have used a new approach to identifying effective areas in IoT. In this research, we used web mining approaches to extract knowledge from the content of web pages. </span></p><p class="c23"><span class="c2">The web mining techniques are those used in the data mining of web documents for automotive knowledge discovery (Etzioni 1996). Web mining, as a multi-disciplined activity, employs techniques from some areas such as information retrieval, statistics, machine learning, and natural language processing. Moreover, it is divided into three main categories, including &ldquo;web content mining&rdquo;, &ldquo;web usage mining&rdquo; and &ldquo;web structure mining&rdquo;. The data mining methods are used for the &ldquo;web content mining&rdquo; of unstructured and semi-structured web pages. The &ldquo;web usage mining&rdquo; focuses on the interactions of users and servers. The &ldquo;web structure mining&rdquo; utilizes hyperlinks as sources of information (Etzioni 1996). As the goal of this paper is to analyze text documents, the study is classified as a &ldquo;web content mining&rdquo; one. It is noteworthy that there is a high similarity between web content mining and text mining. Thus, text mining techniques can be used in the field of web content mining. In order to analyze the text contents of web pages, the co-word analyzing method along social network analysis (SNA) is applied in this paper.</span></p><h1 class="c25 c45"><span class="c2">2. literature review</span></h1><h2 class="c24"><span class="c2">2.1 IoT</span></h2><p class="c23"><span class="c2">In 1999, IoT was first introduced by Kevin Ashton at the Massachusetts Institute of Technology (MIT). IoT is defined as a dynamic global network infrastructure with the capability of self-configuring based on the standard and interoperable communication protocols. In IoT, virtual and physical objects have their own identities and attributes. They both can use intelligent interfaces to form an integrated information network (Vermesan 2009). IoT can behave as a superset for connecting devices identified by Near Field Communication (NFC) techniques (S. Li et al. 2015). It has been about twenty years since the invention of IoT, yet, devoid of a flawless intellectual structure for it is almost tangible. In addition, most of the studies took place to solve the problem, have been done only on the online scientific paper databases.</span></p><h2 class="c24"><span class="c2">2.2 Social Network Analysis and Co-word Analysis</span></h2><p class="c23"><span class="c2">Research methodology of SNA is based on finding actors relationships. Actors can refer to individuals, organizations, or events which can be our nodes in a graph, and the relationships between nodes are represented by edges or links. The SNA is utilized to analyze the graph structure and relationships between the actors (Wasserman and Faust 1994; Borgatti et al. 2002). In this paper, the keywords and key phrases are considered as nodes, and the co-occurrence of keywords and key phrases in the same paragraph are specified as edges. In other words, if the keyword A and keyword B have appeared in the same paragraph, an edge is created between them. This co-occurrence is known as the co-word analysis.<br>The co-word analysis is known as a quantitative method for disclosing scientific relationships (Q. He 1999). It is used by many researchers to evaluate the research and development effects (Kostoff 1993) and to analyze the study trends (Dehdarirad et al. 2014). The main co-word analysis features are related to the visualization of a specific area&rsquo;s intellectual structure and to design a conceptual map (Ding et al. 2001). In this method, a co-word occurrence matrix is used as the input (Salton and McGill 1983; Ronda&#8208;Pupo and Guerras&#8208;Martin 2012). The keywords or key phrases are inserted in the first column and row of this matrix. Their co-occurrence frequency in the same paragraph is presented by a number in the cell where the keyword or key phrase in the first row crosses the other one from the first column. The co-word analysis has been mostly applied on databases such as WOS while there is a lot of more information all over the web. In such databases, keywords and key phrases are determined by authors while determining them in web pages is unusual and rare.</span></p><h2 class="c20"><span class="c2">2.3 Related Studies</span></h2><p class="c23"><span class="c2">Most resources for studying the field of co-word analysis are related to databases in which the keywords and key phrases are provided by their authors. Such studies have less complexity when compared to investigating web documents. Moreover, the web contents are more unstructured than databases of scientific articles. The co-word analysis has been used in many fields such as tourism (de la Hoz-Correa et al. 2018), recommender systems (Hu and Zhang 2015), marketing (Wang et al. 2015), social networks (Gan and Wang 2015), and cancer research (Xie 2015). In the following section, applications of the co-word analysis are surveyed in more details.</span></p><p class="c23"><span class="c2">In order to visualize the intellectual structure of the economic transition effect on the Central Eastern European companies, a co-word analysis was utilized during 1989&ndash;2013. The study was performed on 2053 scholar papers collected from the most comprehensive management databases. It was proven that the co-word analysis is an appropriate method for quantifying and visualizing the intellectual structure of the transition. Lastly, the most influential factors affecting companies were identified (Topalli and Ivanaj 2016). Ravikumar et al. (Ravikumar et al. 2015) for getting the intellectual structure of scientometrics in the period of time 2005&ndash;2010, studied on 959 full-text articles by text mining and co-word analysis. In another paper, in order to investigate the technological energy trends of hydrogen, a co-word analysis was employed. According to the study, the trends are focused on converting hydrogen applications. Moreover, three main sub-patterns of production, storage, and conversion were identified in the network (Wu and Leu 2014). Olmeda-G&oacute;mez et al. (Olmeda-G&oacute;mez et al. 2017) analyzed Spanish institutions researchers&rsquo; papers that published under the category &lsquo;Information Science &amp; Library Science&rsquo; by co-keyword and document co-citation. By Co-citation analysis, they identified nine research fronts: digital rights management, citation analysis, translation services, bibliometric analysis, co-authorship, electronic books, webometrics, information systems, and worldwide web. Due to the abrupt changes in China&#39;s urbanization and a large amount of published papers in the field a dynamic co-word analysis was carried out on 4449 papers from the WOS database during the 1998&ndash;2015 (Zhang et al. 2017). Another approach, based on the co-word analysis, clustering, and SNA, was taken in order to visualize the intellectual structure for doctoral dissertations in the Chinese libraries during 1994&ndash;2011. The co-word analysis was mapped in two positions of binary and weighted graphs. Lastly, 12 clusters were found and labeled (Zong et al. 2013). For studying the evolution of research and identifying new emergence topics, Hou et al. (Hou et al. 2018) collected 7574 articles published in 10 information science (IS) journals in the period 2009-2016. By document co-citation and comparing to their previous study on documents that collected until 2008, they showed significant changes had occurred in the research topics in the IS domain.</span></p><h1 class="c25 c40"><span class="c2">3. Methodology</span></h1><p class="c23"><span class="c2">This study tries to explain the methodology and the challenges of studying different areas of IoT using 100 best blogs on the field. For this, the SNA techniques, especially the co-word analysis, are used in this research. The methodology is divided into five stages, including data collection, structuring, preprocessing, web mining, and finally identifying effective areas in the IoT domain.</span></p><h2 class="c24"><span class="c2">3.1 Data Collection</span></h2><p class="c23"><span class="c2">The data was gathered from 100 best IoT blogs ranked by the Feedspot website. For doing this, a bot crawler was built to crawl the blogs and catch their target information. The bot crawler was built using Python 3.6 with Selenium 3.10.0 and BeautifulSoup 4.6.0 packages to gather the first and second-level links of the selected blogs. The information sum of 3500 links and totally 92966 paragraphs were collected. In this phase, the challenging issue was related to the blog&rsquo;s non-integrated structure as each blog had its HTML and CSS structures. The designed web crawler would collect all paragraphs in the blogs. Another problem while collecting the data was, the large amount of information interrupting the crawling process, which delayed the procedure. </span></p><h2 class="c24"><span class="c2">3.2 Structuring the Collected Data</span></h2><p class="c23"><span class="c2">&nbsp;The collected data was not structured and comprised of many HTML tags. In this phase, for structuring the data, the regular expression patterns were utilized followed by the Python 3.6&rsquo;s &#39;re&#39; package.</span></p><h2 class="c24"><span class="c2">3.3 Preprocessing</span></h2><p class="c23"><span class="c2">Another essential step to take before the text mining stage is data preprocessing. The N-gram (Broder et al. 1997) was used to detect the key phrases in texts. Before the use of N-gram, the stop words were deleted from the texts, and a Porter stemmer was executed by the NLTK package in Python 3.6. We eventually selected 60 numbers of high-frequency words. (Litterio et al. 2017) applied a model for finding potential influencers in the field of marketing within online communities, which was presented by Scoponi et al. (Scoponi et al. 2016).The model comprised of a two-dimensional matrix based on eigenvector centrality and betweenness centrality. Persons with higher eigenvector centrality and betweenness centrality are considered more potential influencers on communities. Litterio et al. utilized the model for classifying actors into four classes, namely &#39;potential influencers&#39;, &#39;brokers&#39;, &#39;actors with important connections&#39; and &#39;secondary actors&#39;. &#39;Potential influencers&#39; refer to actors with the high value of eigenvector centrality and betweenness centrality. This group has high influence on other groups of actors to make a decision. Actors with high betweenness centrality and low eigenvector centrality are categorized into &#39;brokers&#39;. Because of high betweenness centrality, brokers play role of bridge in a network. Actors with low betweenness centrality and high eigenvector centrality are considered as &#39;actors with important connections&#39;. The final group with low betweenness centrality and low eigenvector centrality is classified into &#39; Secondary actors&#39;. In our study, a matrix model is utilized for identifying the effective areas in the field of IoT. Using the model determines that which one of the communities has more effects in IoT (potential influencers), which one connects different areas together (brokers) and which one links to important fields (actors with important connections). Our proposal is an adaptation of the matrix model presented by Scoponi et al. (2016) to classify the actors of a social network in terms of their level of influence through two complementary metrics.</span></p><h1 class="c40 c25"><span class="c2">4.Results</span></h1><h2 class="c20"><span class="c2">4.1 Interpretation of web mining output</span></h2><p class="c23"><span class="c2">As is shown in Figure 3, there are eight communities found, of which the first, named Security and indicated by purple, is related to the threats and risks. &quot;Security&quot; is the most prominent member of the community. &quot;Cybersecurity&quot; and &quot;risk&quot; are other challenging issues.
